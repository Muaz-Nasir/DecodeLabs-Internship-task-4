```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = "Dataset for Data Analytics.xlsx"

try:
    df = pd.read_excel(path)
    print(f"File loaded successfully! Found {len(df)} rows.")
except FileNotFoundError:
    print(f"Error: Can't find '{path}'. Check your file name in the Colab sidebar.")
    exit()

df["CouponCode"] = df["CouponCode"].fillna("NO_COUPON")
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.drop_duplicates(subset=["OrderID"], keep="first")

sns.set_theme(style="whitegrid")


plt.figure(figsize=(8, 5))

product_counts = df["Product"].value_counts()

sns.barplot(x=product_counts.index, y=product_counts.values, palette="Blues_r")

plt.title("Total Transaction Volume by Product Class", fontsize=14, pad=15)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Number of Orders", fontsize=12)
plt.xticks(rotation=15)

plt.tight_layout()
plt.show()


plt.figure(figsize=(6, 6))

payment_counts = df["PaymentMethod"].value_counts()

plt.pie(
    payment_counts.values,
    labels=payment_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette("Pastel1")
)

plt.title("Proportional Share of Payment Methods", fontsize=14, pad=20)

plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 5))

df["YearMonth"] = df["Date"].dt.to_period("M")
monthly_trend = df.groupby("YearMonth")["OrderID"].count()

months = monthly_trend.index.astype(str)
sales_volume = monthly_trend.values

plt.plot(months, sales_volume, marker='o', color='#2b5c8f', linewidth=2, markersize=6)

plt.title("Operational Sales Transaction Trend Over Time", fontsize=14, pad=15)
plt.xlabel("Timeline (Year-Month)", fontsize=12)
plt.ylabel("Total Transaction Volume", fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
